#  Try to predict sequences using LSTM.
# *The input is a list of size n containing the first n natural numbers and the target sequence is the (n+1)th number.*

```python
import torch
import torch.nn as nn
import random
import torch
import time
```

```python
# Generate the input and target sequence. Index contains the list of target values.
# Data contains the target values mapped to the input tensor.


torch.manual_seed(0)  # Set a seed for reproducibility

index = []
data = {}
for _ in range(300):
    tensor_length = random.randint(1, 100)
    tensor_values = [[i ] for i in range(1, tensor_length + 1)]
    index.append(tensor_length+1)
    data[tensor_length+1] = torch.tensor(tensor_values, dtype=torch.float32)

print(data[index[0]])
print(index[0])
print(len(data))
print(len(index))

data[index[0]].shape
```

```python
# Hidden_dim =32,lr=0.001,num_epochs=100 and MAE as loss function were giving best results. 

class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(LSTMModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.lstm = nn.LSTM(input_dim, hidden_dim)
        self.fc = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, input_seq):
        lstm_out, _ = self.lstm(input_seq.view(len(input_seq), 1, -1))
        predictions = self.fc(lstm_out.view(len(input_seq), -1))
        return predictions[-1]

# Set the hyperparameters
input_dim  = 1  # Input dimension
hidden_dim = 32 # Hidden layer dimension
output_dim = 1  # Output dimension

# Instantiate the model
model = LSTMModel(input_dim, hidden_dim, output_dim)
print(model)

# Define loss function and optimizer
#criterion = nn.MSELoss()
criterion = nn.L1Loss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```
